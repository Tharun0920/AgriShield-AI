```python
import os
import cv2
import matplotlib.pyplot as plt

# Point to your images folder
image_dir = '../data/images/'

# Get the names of all the disease folders
disease_classes = os.listdir(image_dir)
print(f"Found {len(disease_classes)} different disease classes.")

# Let's open the very first folder and look at the very first image
first_disease = disease_classes[0]
first_image_name = os.listdir(os.path.join(image_dir, first_disease))[0]
image_path = os.path.join(image_dir, first_disease, first_image_name)

# Read and display the image
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Fix colors

plt.imshow(img)
plt.title(f"Sample Disease: {first_disease}")
plt.axis('off')
plt.show()
```
